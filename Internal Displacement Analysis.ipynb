These are countries with natural disasters that cause a good amount of people in the countries to be displaced form their homes, sometimes temporarily, and sometimes permanently. 

People who leave their countries of residence are not regarded as internally displaced poeple but refugees (in the new places they stay)


The dataset has the following columns; 

1. **district**: The name of the district (or administrative division) within a country. This helps in identifying the location of the data.

2. **elevation**: The height of the district above sea level, typically measured in meters. Higher elevations may be more resistant to flooding but can experience other climate conditions.

3. **Area**: The total land area of the district, often measured in square kilometers. This can give an idea of the size of the district.

4. **lat (Latitude)**: The geographic coordinate specifying the north-south position of a point on the Earth's surface, expressed in degrees.

5. **lng (Longitude)**: The geographic coordinate specifying the east-west position of a point on the Earth's surface, expressed in degrees.

6. **Year**: The year of the recorded data. This can help in understanding trends over time, particularly for events like disasters or population growth.

7. **Disaster Internal Displacements**: The number of people internally displaced due to natural disasters within the district. This often occurs due to hazards like flooding, droughts, or storms.

8. **Hazard Type**: The type of natural disaster or hazard that affected the district, such as floods, droughts, or storms.

9. **total_precip**: The total precipitation (rainfall) in the district, typically measured in millimeters (mm) or inches, during the period of observation. Precipitation is crucial for understanding water availability and the likelihood of flooding.

10. **drought_index**: Total number of days in the year with 0 inches rainfall

11. **total solar**: The total solar radiation received in the district, typically measured in kilowatt-hours per square meter. This is important for understanding climate and energy potential.

12. **heavy rain**: Total number of days in a year with heavy rainfall

13. **storm days**: The number of days the district experienced storms, including windstorms, thunderstorms, or other severe weather events.

14. **extreme temp**: number of days with extremely high temperature
15. **severerisk**: number of days at sever risk

16. **low_pressure_days**: The number of days with low atmospheric pressure, which can be associated with storms or adverse weather conditions.

17. **pressure_anomaly**: A measure of how much the atmospheric pressure deviates from the normal range. This can indicate unusual weather patterns, such as approaching storms.

18. **Country / Territory**: The name of the country or territory where the district is located.

19. **population**: The number of people living in the district, used for demographic analysis and understanding the impact of natural disasters.

20. **Count of Disasters**: The total number of recorded natural disasters that have affected the district over the period of observation.

21. **pdensity (Population Density)**: The number of people per square kilometer in the district. Population density is useful for assessing the potential human impact of disasters.

22. **distancetowater**: The distance from the district to the nearest water body (river, lake, or reservoir), typically measured in kilometers. This is important for understanding the flood risk.

23. **distancetowater1**: The distance to the second nearest water body.

24. **distancetowater2**: The distance to the third nearest water body.

25. **Country Code**: A standardized code representing the country, often in ISO 3166-1 alpha-3 format (three-letter code).

26. **gdp (Gross Domestic Product)**: The economic output of the district, often estimated from country-level GDP. This indicates the wealth or economic capacity of the district.

3.1 Objective: Understand how the displacement varies between countries and years. 

To do this, we will first group the dataset by countries using the .groupby function and sunm all the displacements per country and plot a barplot using the pandas .plot function with the kind='bar' parameter to indictate bar plot

After grouping the data , the new dataframe would be aggregate sum of total displacements per country, The sort_values is to sort the dataset , the default setting is to sort it in ascending order , that is , From lowest to highest, thus the ascending=False parameter which would arrange it in Descending order (highest to lowest)
After which we would be resetting the index to make it come out as a dataframe while assigning new names to the columns

```python
displacement_by_country = df.groupby('Country / Territory')['Disaster Internal Displacements'].sum().sort_values(ascending=False)

displacement_by_country = displacement_by_country.reset_index()

displacement_by_country.columns = ['Country / Territory', 'Total_Displacements']


```

```python
# PLot the Bar plot using pandas' built-in plotting
displacement_by_country.plot(x='Country / Territory', y='Total_Displacements', kind='bar', figsize=(10, 6), color='skyblue')

plt.xticks(rotation=0) # Rotation is for better readability
plt.xlabel('Country / Territory')
plt.ylabel('Total Displacements')
plt.title('Total Internal Displacements by Country')

plt.show()

```

The country with the the most internal displacements is Somalia with over 2 million displacements followed by Uganda and Central African Republic from this Dataset. 
Papua New Guniea has the least internal displacements. 


Secondly, we will group the dataset per year and plot a line plot using the pandas plotting function while making the plot better to read with some customizations
like,
 Grid Lines: Both major and minor grid lines are included, making it easier to read the graph. The grid lines are styled with dashed lines for a clearer visual distinction.

```python
# PLot the Bar plot 
displacement_by_year = df.groupby('Year')['Disaster Internal Displacements'].sum()

displacement_by_year = displacement_by_year.reset_index() # reset the index to get a new Df

displacement_by_year.columns = ['Year','Total_Displacements']
```

```python
displacement_by_year.plot(x='Year',y='Total_Displacements',kind='line',marker='o',figsize=(10,6),title='Total Internal Displacement Over the Years')

plt.xticks(ticks=displacement_by_year['Year'])
plt.grid(which='both', linestyle='--', linewidth=0.5)
plt.show()
```

1. Sharp Increase in 2023: There is a dramatic spike in total internal displacements in 2023. The displacement has jumped significantly compared to all previous years, indicating a major event or series of events that triggered large-scale internal displacement in that year.
2. Relatively Stable Pattern Before 2023: Between 2008 and 2022, the total displacements were fairly low, with minor fluctuations. The values remain under 0.3 million(300,000) displacements until the sharp increase in 2023.
3. Gradual Growth from 2013 to 2018: There is a gradual increase in internal displacements from 2013 to around 2018, followed by a period of relative stability or minor decreases until 2022.

3.2  Objective: Identify features that have strong relationships with internal displacement. 

In this step, we will be analyzing the variables in the dataset that would have impact on the target variable("DISASTER INTERNAL DISPLACEMENTS" ) 

To know them we'll be calculating the correlation coeffictient of the numercial variables as regards to the distaster internal displaceents 

the pandas .corr() function would be used for that. 
The Pearson's coefficient, also known as Pearson's correlation coefficient is a statistical measure that quantifies the strength and direction of the linear relationship between two continuous variables. Its value ranges from -1 to 1: 
A value of 1 indicates a perfect positive linear relationship, meaning that as one variable increases, the other variable also increases proportionally.
A value of -1 indicates a perfect negative linear relationship, meaning that as one variable increases, the other variable decreases proportionally.
A value of 0 indicates no linear correlation between the variables.
Pearson's coefficient is calculated by dividing the covariance of the two variables by the product of their standard deviations. It's important to note that while it measures linear relationships, it may not accurately represent nonlinear relationships between variables. Additionally, Pearson's coefficient is sensitive to outliers, which can significantly affect its value.



```python
correlation_matrix = df.corr(method='pearson',numeric_only=True)

print(correlation_matrix['Disaster Internal Displacements'].sort_values(ascending=False))
```

1. Severe Risk and Count of Disasters have the highest positive correlation with internal displacements, indicating that areas with higher disaster risks and more frequent disasters experience more displacements.
2. Extreme temperatures and solar anomalies also show a positive, though weaker, relationship with displacements, suggesting that climate-related factors may play a role in triggering internal migrations.
3. GDP and elevation have the strongest negative correlations with internal displacements, meaning that wealthier regions or those at higher elevations tend to experience fewer displacements.
4. Population density, storm days, and heavy rain have weak negative correlations, implying that other factors may play a larger role in causing displacement than population or rainfall intensity.

We will be plotting a scatter plot to check trends in relationships between the variables with the strongest correlations like Severrisk, Count of disatsers, GDP, and elevation

3.3 Objective: Analyze trends in displacement for each country over the years.

We will be doing this by grouping the dataset per country and year and summing the total internal displacments and plot a line plot to see how disaster internal displacements changed over time in each country 

The unstack() function pivots the grouped data, transforming the grouped data from a multi-level index (with both Country and Year) into a DataFrame format where the Year values become the columns. This means the dataframe will have one row for each Country, and each column will represent a different Year. The values in the DataFrame will be the total displacements for that country in that year. 

```python
displacement_over_time = df.groupby(['Country / Territory','Year'])['Disaster Internal Displacements'].sum().unstack()

displacement_over_time
```

Next is to plot the line plot , but before we do that , we will be transposing the dataframe to make it easy to plot the line chart

By using .T, the DataFrame is transposed so that Year becomes the index (rows), and Country becomes the columns. 

```python
# Creating the line plot

displacement_over_time.T.plot(kind='line', marker='o', linestyle='-', title='Internal Displacement Over Time by Country',figsize=(8,6))

plt.xlabel('Year')
plt.ylabel('Total Internal Displacement')
plt.title('Internal Displacement Over Time by Country', fontsize=16)
plt.xticks(rotation=45)  
plt.grid(True)  
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')  # Place legend outside the plot area

# Show the plot
plt.tight_layout()  # Adjust layout to make room for the legend
plt.show()
```

From the line plot, we can see ;

1. Somalia's Sharp Increase: There is a significant surge in internal displacement in Somalia around 2021-2022, reaching over 2 million, which is much higher than the other countries.

2. Stable Trends for Most Countries: Countries like Zimbabwe, Uganda, Central African Republic, and Papua New Guinea show relatively stable or moderate fluctuations over the years without any drastic spikes like Somalia.

3. Central African Republic: While it has some fluctuations, the internal displacement figures remain relatively low and steady compared to Somalia.

4. Somalia's Dominance in Recent Years: The plot shows that Somalia's internal displacement figures dwarf those of other countries in the dataset, particularly after 2020.

We will Aggregate displacement by hazard type to see which disasters have the largest impact using the mean.

3. Moderate Impact from Storms and Earthquakes: Storms and earthquakes are also notable contributors, though their average displacement is much lower compared to floods and droughts, reflecting their occurrence or intensity in the region.
